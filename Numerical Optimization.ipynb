```python
def rosenbrock_func(x):
    """
    Calcualtes the Rosenbrock function value for a given point.
    
    Args:
        x: the two dimensional point at which the function value should be calculated.
    Returns:
        the Rosenbrock function value for a given point
    """
    x_1, x_2 = x
    return 100*(x_2 - x_1**2)**2 + (1-x_1)**2


def backtracking_line_search(func, x_k, p, rho=0.5, c=1e-4,
                             alpha=1, method='steepest descent', verbose=True):
    """
    Calcualtes the step length parameter value parameter for a line search.
    
    Args:
        func: the function to be optimized.
        x_k: the k-th iterate value for the function argument.
        p: the descent direction.
        rho: contraction factor.
        c: constant in range (0, 1).
        alpha: initial value of the step length parameter.
        method: a string value that indicates the type of method to be used. 
                Can be set to 'steepest descent' or 'newton'.
        verbose: a boolean parameter to print the values of alpha if 
                 it is set to True else no.
    Returns:
        the value of the step length that satisfies the armijo condition.
    """
    grad_k = rosenbrock_grad(x_k)
    descent = np.dot(grad_k, p(x_k, method=method))
    while func(x_k + alpha*p(x_k)) >= func(x_k) + c*alpha*descent:
        alpha = rho*alpha
        if verbose:
            print(alpha)
    alpha_k = alpha
    return alpha_k


def rosenbrock_grad(x):
    """
    Calcualtes the gradient of the Rosenbrock function.
    
    Args:
        x: the two dimensional point at which the gradient value should be calculated.
    Returns:
        the gradient of the Rosenbrock function at the given point.
    """
    x_1, x_2 = x
    return np.array([-400*x_1*(x_2-x_1**2)-2*(1-x_1),
                     200*(x_2-x_1**2)])


def rosenbrock_hessian(x):
    """
    Calcualtes the hessian of the Rosenbrock function.
    
    Args:
        x: the two dimensional point at which the hessian value should be calculated.
    Returns:
        the hessian of the Rosenbrock function at the given point.
    """
    x_1, x_2 = x
    return np.array([[-400*x_2 + 1200*x_1**2 + 2, -400*x_1],
                     [-400*x_1, 200]])


def p(x, method='steepest descent'):
    """
    Calcualtes the descent direction for the Rosenbrock function.
    
    Args:
        x: the two dimensional point at which the descent value should be calculated.
        method: a string value that indicates the type of method to be used. 
                Can be set to 'steepest descent' or 'newton'.
    Returns:
        the descent direction for the Rosenbrock function for the given method.
    """
    grad = rosenbrock_grad(x)
    if method == 'steepest descent':
        nrm = np.linalg.norm(grad)
        return -1*grad / nrm
    if method == 'newton':
        hessian = rosenbrock_hessian(x)
        return -np.dot(np.linalg.inv(hessian), grad)


def descent(func, p, x_0, alpha, n_iter, method):
    """
    Optimizes the given function using line search methods.
    
    Args:
        func: the function that should be optimized.
        p: the descent direction.
        x_0: the two dimensional initial point at which the descent value should be calculated.
        alpha: the initial value for the step length parameter.
        n_iter: the number of iterations the algorithm should run.
        method: a string value that indicates the type of method to be used. 
                Can be set to 'steepest descent' or 'newton'.
    Returns:
        list of function arguments generated while optimizing the function and the list of 
        function values evaluated at these arguments.
    """
    f = []
    x = []
    x_old = x_0
    f_old = func(x_0)
    x_new = x_old + alpha*p(x_old)
    f_new = func(x_new)

    f.append(f_old)
    f.append(f_new)
    x.append(x_old)
    x.append(x_new)
    for _ in range(n_iter):
        f_new = func(x_new)
        x_old = x_new
        alpha = backtracking_line_search(func, x_old, p,
                                         alpha=alpha, method=method)
        x_new = x_old + alpha*p(x_old)
        f.append(f_new)
        x.append(x_new)
    return x, f

x_0 = np.array([1.2, 1.2])
x_values, fun_values = descent(rosenbrock_func, p, x_0, alpha=1,
                               n_iter=5000, method='steepest descent')
```

```python
def rosenbrock_func(x):
    """
    Calcualtes the Rosenbrock function value for a given point.
    
    Args:
        x: the two dimensional point at which the function value should be calculated.
    Returns:
        the Rosenbrock function value for a given point
    """
    x_1, x_2 = x
    return 100*(x_2 - x_1**2)**2 + (1-x_1)**2


def backtracking_line_search(func, x_old, x_new, B, p, rho=0.5, 
                             c=1e-4, alpha=1, verbose=True):
    """
    Calcualtes the step length parameter value parameter for a line search.
    
    Args:
        func: the function to be optimized.
        x_old: the (k-1)-th iterate value for the function argument.
        x_new: the (k)-th iterate value for the function argument.
        B: the hessian approximation matrix.
        p: the descent direction.
        rho: contraction factor.
        c: constant in range (0, 1).
        alpha: initial value of the step length parameter.
        verbose: a boolean parameter to print the values of alpha if 
                 it is set to True else no.
    Returns:
        the value of the step length that satisfies the armijo condition.
    """
    grad_k = rosenbrock_grad(x_old)
    while func(x_old + alpha*p(x_old)) >= func(x_old) + c*alpha*np.dot(grad_k, p(x_old, x_new, B)):
        alpha = rho*alpha
        if verbose:
            print(alpha)
    alpha_k = alpha
    return alpha_k


def rosenbrock_grad(x):
    """
    Calcualtes the gradient of the Rosenbrock function.
    
    Args:
        x: the two dimensional point at which the gradient value should be calculated.
    Returns:
        the gradient of the Rosenbrock function at the given point.
    """
    x_1, x_2 = x
    return np.array([-400*x_1*(x_2-x_1**2)-2*(1-x_1),
                         200*(x_2-x_1**2)])


def BFGS(x_old, x_new, B):
    """
    Calcualtes the hessian approximation matrix.
    
    Args:
        x_old: the (k-1)-th iterate value for the function argument.
        x_new: the (k)-th iterate value for the function argument.
        B: the hessian approximation matrix.
    Returns:
        the updated value of the hessian approximation matrix using BFGS algorithm.
    """
    x_1, x_2 = x
    s_k = x_new - x_old
    y_k = rosenbrock_grad(x_new) - rosenbrock_grad(x_old)
    numer = np.dot((y_k - B@s_k), (y_k - B@s_k).T)
    denom = np.dot((y_k - B@s_k).T, s_k)
    B_new = B + numer / denom
    return B_new
    

def p(x_old, x_new, B):
    """
    Calcualtes the descent direction for the Rosenbrock function.
    
    Args:
        x_old: the (k-1)-th iterate value for the function argument.
        x_new: the (k)-th iterate value for the function argument.
        B: the hessian approximation matrix.
    Returns:
        the descent direction for the Rosenbrock function.
    """
    grad = rosenbrock_grad(x_old)
    hessian_approx = BFGS(x_old, x_new, B)
    return -np.dot(np.linalg.inv(hessian_approx), grad)


def descent(func, p, x_0, B, alpha, n_iter):
    """
    Optimizes the given function using line search methods.
    
    Args:
        func: the function that should be optimized.
        p: the descent direction.
        x_0: the two dimensional initial point at which the descent value should be calculated.
        B: the hessian approximation matrix.
        alpha: the initial value for the step length parameter.
        n_iter: the number of iterations the algorithm should run.
    Returns:
        list of function arguments generated while optimizing the function and the list of 
        function values evaluated at these arguments.
    """
    f = []
    x = []
    x_old = x_0
    f_old = func(x_0)
    x_new = x_old + alpha*p(x_old)
    f_new = func(x_new)
    
    f.append(f_old)
    f.append(f_new)
    x.append(x_old)
    x.append(x_new)
    for _ in range(n_iter):
        f_new = func(x_new)
        alpha = backtracking_line_search(func, x_old, x_new, B, 
                                         p, alpha = alpha)
        x_old = x_new
        x_new = x_old + alpha*p(x_old)
        f.append(f_new)
        x.append(x_new)
    return x, f
        
x_0 = np.array([1.2, 1.2])
x_values, fun_values = descent(rosenbrock_func, p, x_0, alpha=1, 
                               n_iter=5000)
```

```python
def cholesky(A):
    """
    Performs a Cholesky decomposition of given matrix.

    Args:
        A: a symmetric, positive definite matrix.
    Returns:
        a Cholesky decomposition of A.
    """
    n = len(A)

    # Create zero matrix for L
    L = [[0.0] * n for i in range(n)]

    # Perform the Cholesky decomposition
    for i in range(n):
        for k in range(i+1):
            tmp_sum = sum(L[i][j] * L[k][j] for j in range(k))
            
            if (i == k): # Diagonal elements
                L[i][k] = np.sqrt(A[i][i] - tmp_sum)
            else:
                L[i][k] = (1.0 / L[k][k] * (A[i][k] - tmp_sum))
    return L

# A = [[6, 3, 4, 8], [3, 6, 5, 1], [4, 5, 10, 7], [8, 1, 7, 25]]
np.array(cholesky(A))
```

```python
def get_r(x, radius, B, g):
    """
    Computes the minimizer of the model.

    Args:
        x: the point at which the function gradient and hessian are evaluated.
        radius: trust-region radius.
        B: the hessian of the function.
        g: the gradient of the function.
    Returns:
        the unconstraint minimizer of the model or 1.
    """
    denom = g.T@B@g
    if denom <= 0:
        r = 1
    else:
        r = min(1, np.linalg.norm(g)**3 / (radius*denom))
    return r


def dogleg(x, radius, B, g):
    """
    Performs a dogleg algorithm to find
    the descent direction which minimizes the model.

    Args:
        x: the input point.
        radius: radius of the trust-region.
        B: the hessian of the model.
        g: the gradient of the model
    Returns:
        the descent direction that minimizes the model at the given point x.
    """
    p_U = -(g.T@g / g.T@B@g) * g
    p_B = -np.linalg.inv(B)@g
    r = get_r(x, radius, B, g)
    if (r >= 0) and (r < 1):
        p_hat_r = r*p_U
    elif (r >= 1) and (r <= 2):
        p_hat_r = p_U + (r-1)(p_B - p_U)
    return p_hat_r


def m(f, x, p, B, g):
    """
    Computes the model approximation value of the true function.

    Args:
        f: the function to be optimized.
        x: the input value where the function,
           its gradient and hessian are evaluated.
        p: the descent direction.
        B: the hessian of the function f.
        g: the gradient of the function f.
    Returns:
        the value of the second order approximation of f evaluated at x.
    """
    return f(x) + g@p + 1/2*p.T@B@p


def trust_region(f, m, x, upper_radius, eta, n_iter=100):
    """
    Optimizes the given function using trust-region methods and
    computing the descent direction by the dogleg algorithm.

    Args:
        f: the function to be optimized.
        m: second order approximation of the true function.
        x: the input value where the function,
           its gradient and hessian are evaluated.
        upper_radius: the initial highest value of the radius for trust-region.
        eta: a constant in the range (0,1/4]
        n_iter: number of iterations the algorithm will run.
    Returns:
        list of function arguments generated while
        optimizing the function and the list of
        function values evaluated at these arguments.
    """
    radius = np.random.randint(upper_radius)
    F_x = []
    X = []
    X.append(x)
    F_x.append(f(x))
    for k in range(n_iter):
        # find the descent direction by dogleg method
        B = rosenbrock_hessian(x)
        g = rosenbrock_grad(x)
        p_B = -np.linalg.inv(B)@g
        if np.linalg.norm(p_B) <= radius:
            p = p_B
        elif np.linalg.norm(p_B) / 20 > radius:
            p = -radius * g / np.linalg.norm(g)
        else:
            p = dogleg(x, radius, B, g)
        # compute the next iterate point
        rho = (f(x) - f(x + p)) / (f(x) - m(f, x, p, B, g))
        if rho < 1/4:
            radius = radius / 4
        elif (rho > 3/4) and (np.linalg.norm(rho) == radius):
            radius = min(2*radius, upper_radius)
        else:
            radius = radius
        if rho > eta:
            x = x + p
        else:
            x = x
        X.append(x)
        F_x.append(f(x))
    return X, F_x

x = np.array([-1.2, 1])
trust_region(rosenbrock_func, m, x, upper_radius=5, eta=0.05, n_iter=23)
```

```python
from scipy.linalg import hilbert

def conjugate_gradient(A, b, x_cur, n):
    r_cur = A@x_cur - b
    p_cur = -r_cur
    x_new = 0
    r_new = 0
    for k in range(n+1):
        a = (r_cur.T@r_cur) / (p_cur.T@A@p_cur)
        x_new = x_cur + a*p_cur
        r_new = r_cur + a*A@p_cur
        b_cur = (r_new.T@r_new) / (r_cur.T@r_cur)
        p_new = -r_new + b_cur*p_cur

        p_cur = p_new
        r_cur = r_new
        x_cur = x_new
    return x_cur, r_cur

X = []
R = []
for n in [5, 8, 12, 20]:
    A = hilbert(n)
    b = np.ones(n)
    x_cur = np.zeros(n)
    x,r = conjugate_gradient(A, b, x_cur, n)
    X.append(x)
    R.append(r)
```
